```python
import numpy as np
docs=['go india','india india','hip hip hurray','jeetega bhai jeetega india jeetega','bharat mata ki jai','kohli kohli','sachin sachin','dhoni dhoni','modi ji ki jai','inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences=tokenizer.texts_to_sequences(docs)
sequences
```

```python
from tensorflow.keras.utils import pad_sequences
sequences=pad_sequences(sequences,padding='post')
sequences
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding
model=Sequential()
model.add(Embedding(input_dim=18, output_dim=2, input_length=5))
model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred=model.predict(sequences)
print(pred)
```
